```python
#2. Afficher le résumer statistique des deux Data Sets avec 
# une interprétation des résultats obtenues.

import pandas as pd

# Charger les Data Sets
datasetExpSal = pd.read_csv("Salary_Data.csv")
datasetAssurance=pd.read_csv("insurance.csv")

# Résumé statistique du datasetExpSal 
print("\nRésumé statistique du Salary Data:")
print(datasetExpSal.describe())

# Résumé statistique du datasetAssurance
print("\nRésumé statistique d'Insurrance Data :")
print(datasetAssurance.describe())
```

```python
#3. Afficher la nuage des points du premier data set « Expérience / Salaire » en utilisant
#matplotlib et pandas.

import matplotlib.pyplot as plt 

file_salary = ["Salary_Data.csv"]
column_names_salary = [["years", "salary"]]
# Nuage de points de Salary Data 
df1 = pd.read_csv(file_salary[0], names=column_names_salary[0])
df1.plot.scatter(x="years", y="salary", color="DarkGreen")
plt.xlabel("Année d'expérience")
plt.ylabel("Salaire")
plt.show()
```

```python
from pandas.plotting import scatter_matrix

# Nuage de points d'Insurrance Data 
scatter_matrix(datasetAssurance[['age', 'bmi', 'children', 'charges']],
               alpha=0.8, figsize=(8,8), diagonal='hist')
plt.suptitle('Nuages de points - Features de Insurrance Data', y=0.95)
plt.show()
```
