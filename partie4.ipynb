```python
#1. en utilisant l’API sklearn entraîner le modèle 
# par intermédiaire de algorithme de la régression
#linéaire et puis la régression linière polynomiale 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import pandas as pd

# Charger les données
data_set = pd.read_csv("china_gdp.csv")  
X = data_set['Year'].values.reshape(-1,1)
y = data_set['Value'].values


# Créer le modèle de régression linéaire
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Créer le modèle de régression polynomiale
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Prédire les données de test
y_pred_lin = lin_reg.predict(X)

# Prédire les données du set de test pour le modèle polynomiale
y_pred_poly = poly_reg.predict(X_poly)

# Créer un DataFrame avec les données
df = pd.DataFrame({'X': X.flatten(), 'y': y, 'Linear Regression': y_pred_lin, 'Polynomial Regression': y_pred_poly})

# Afficher le tableau
print(df)

```

```python
import matplotlib.pyplot as plt

# Visualiser les résultats
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred_lin, color='green', label='Linear Regression')
plt.plot(X, y_pred_poly, color='red', label='Polynomial Regression')
plt.legend()
plt.show()

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np


# Calculer les métriques d'évaluation pour le modèle linéaire
mse_linear = mean_squared_error(y, y_pred_lin)
rmse_linear = np.sqrt(mse_linear)
mae_linear = mean_absolute_error(y, y_pred_lin)

# Calculer les métriques d'évaluation pour le modèle polynomiale
mse_poly = mean_squared_error(y, y_pred_poly)
rmse_poly = np.sqrt(mse_poly)
mae_poly = mean_absolute_error(y, y_pred_poly)

# Afficher les résultats
print("Régression linéaire:")
print(f"Mean Squared Error (MSE): {mse_linear}")
print(f"Root Mean Squared Error (RMSE): {rmse_linear}")
print(f"Mean Absolute Error (MAE): {mae_linear}")
print("\nRégression polynomiale:")
print(f"Mean Squared Error (MSE): {mse_poly}")
print(f"Root Mean Squared Error (RMSE): {rmse_poly}")
print(f"Mean Absolute Error (MAE): {mae_poly}")
```
